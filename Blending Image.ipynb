```python
import numpy as np
import cv2
%matplotlib inline 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python

img_water = mpimg.imread('backg2.jpg')
resized = cv2.resize(img_water, (300, 300))
plt.imshow(resized)
plt.title('Image')
plt.show()
```

```python
img_horse = mpimg.imread('model.jpg')
resized2 = cv2.resize(img_horse, (300, 300)) 

plt.imshow(resized2)
plt.title('Image')

plt.show()
```

```python
blend = cv2.addWeighted(resized,0.5,resized2,0.6,0)
plt.imshow(blend)
plt.title('Blended Image')
plt.show()
```
